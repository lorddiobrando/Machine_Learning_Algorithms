```python
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creating artificial data: F(x) = 2.5x + 6
x = np.array([[]])
y = np.array([])
for i in range(1, 251, 2):
  x = np.append(x, [[1, i]])
  y = np.append(y, 2.5 * i + 6)
x
```

```python
x = x.reshape(int(len(x) / 2), 2)
x
```

```python
y = y.reshape(len(y), 1)
y
```

```python
psuedo_inverse = np.linalg.pinv(x)
psuedo_inverse
```

```python
w = np.matmul(psuedo_inverse, y)
w
```

```python
y_hat = np.matmul(x, w)
y_hat
```

```python
plt.plot(x[:, 1], y_hat)
plt.scatter([i for i in x[:, 1] if not i % 3], [y[ind] for ind, i in enumerate(x[:, 1]) if not i % 3])
plt.show()
```

```python
# F(x) = 2.5 * x + 6 + random(0...20)
y_rand = np.array([])
for i in range(1, 251, 2):
  rand = random.random()
  y_rand = np.append(y_rand, 2.5 * i + 6 + rand * 20)
```

```python
y_rand = y_rand.reshape(len(y), 1)
y_rand
```

```python
w = np.matmul(psuedo_inverse, y_rand)
w
```

```python
y_hat_rand = np.matmul(x, w)
y_hat_rand
```

```python
plt.plot(x[:, 1], y_hat_rand)
plt.scatter([i for i in x[:, 1] if not i % 3], [y_rand[ind] for ind, i in enumerate(x[:, 1]) if not i % 3])
plt.show()
```

```python
df = pd.read_csv("Iris.csv")
df
```

```python
augment = [1 for i in range(150)]
df.insert(0, "Augmentation", augment)
df
```

```python
temp = []
for i in range(len(df)):
  if df["Species"][i] == "Iris-setosa":
    temp.append(1)
  else:
    temp.append(-1)
df["Label"] = temp
df.head(150)
```

```python
df.drop(["Species", "Id"], axis = 1, inplace = True)
df.head(51)
```

```python
for i in range(50, len(df)):
  for j in range(6):
    df.iloc[[i], [j]] *= -1
df.tail()
```

```python
df.head(150)
```

```python
columns = [i for i in range(len(df.columns[:-1]))]
x_train = np.array([])
for i in range(0, 101, 50):
  for j in range(40):
    x_train = np.append(x_train, [df.iloc[[j + i], columns]])
```

```python
x_train = x_train.reshape(-1, 5)
print(x_train, len(x_train), x_train.shape)
```

```python
x_test = np.array([])
for i in range(40, 150, 50):
  for j in range(10):
    x_test = np.append(x_test, df.iloc[[i + j], columns])
```

```python
x_test = x_test.reshape(-1, 5)
print(x_test, len(x_test))
```

```python
for i in range(10, len(x_test)):
  x_test[i][:] *= -1
x_test
```

```python
x_pinv = np.linalg.pinv(x_train)
print(x_pinv, x_pinv.shape)
```

```python
y = np.array(df['Label'][:120]).reshape(-1, 1)
print(len(y), y.shape)
```

```python
w = np.matmul(x_pinv, y)
print(w, w.shape)
```

```python
y_hat = np.matmul(x_test, w)
print(y_hat, y_hat.shape)
```

```python
for i in y_hat:
  if i[0] > 0:
    print("Class I")
  elif i[0] < 0:
    print("ClassII")
  else:
    print("Undefined")
```
